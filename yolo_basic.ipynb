```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# YOLO 모델 로드 (사전 학습된 모델)
# model = YOLO("yolov8n.pt")  # YOLOv8 nano 모델
model = YOLO("yolo11x.pt")
FILE_NAME = "hh_dash.jpg"
# 이미지에서 보행자 탐지
results = model.predict(source=FILE_NAME, conf=0.5)

# 이미지 로드
image = cv2.imread(FILE_NAME)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # OpenCV는 BGR 형식이므로 RGB로 변환

# 탐지 결과를 이미지에 그리기
for result in results:
    for box in result.boxes:
        cls = int(box.cls)  # 클래스 ID
        confidence = box.conf[0]  # 신뢰도
        if cls == 0:  # 'person' 클래스
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # 경계 상자 좌표
            # 경계 상자 그리기
            cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)
            # 클래스 이름과 신뢰도 추가
            cv2.putText(
                image,
                f"Person {confidence:.2f}",
                (x1, y1 - 10),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.5,
                (255, 0, 0),
                2,
            )

# 결과 이미지 표시
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis("off")
plt.show()
```
