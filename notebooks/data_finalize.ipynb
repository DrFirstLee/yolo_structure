```python
import pandas as pd
```

```python
file_name ="oneway_fast_20250127-151257.csv"
df_raw =pd.read_csv(f"/app/yolo_structure/results/{file_name}")
df_raw              
```

```python
df_raw.detected.unique()
```

```python
df_raw.event.unique()
```

```python
df_raw.frame_hash.nunique()
```

```python
# 조건: PERSON이 frame_hash 기준으로 5회 연속 등장하는 경우
result = (
    df_raw[df_raw["detected"] == "person"]  # PERSON만 필터링
    .groupby("frame_hash")  # frame_hash 기준 그룹화
    .filter(lambda x: len(x) >= 5)  # 5회 이상 연속 발견된 그룹 필터링
)

# 결과 출력
print(result)
result
```

```python
# 조건: PERSON이 frame_hash 기준으로 5회 연속 등장하는 경우
result = (
    df_raw[df_raw["event"] == "STOP"]  # PERSON만 필터링
    .groupby("frame_hash")  # frame_hash 기준 그룹화
    .filter(lambda x: len(x) >= 3)  # 5회 이상 연속 발견된 그룹 필터링
)

# 결과 출력
print(result)
result
```

```python
# 조건: PERSON이 frame_hash 기준으로 5회 연속 등장하는 경우
result = (
    df_raw[df_raw["event"] == "REVERSE"]  # PERSON만 필터링
    .groupby("frame_hash")  # frame_hash 기준 그룹화
    .filter(lambda x: len(x) >= 5)  # 5회 이상 연속 발견된 그룹 필터링
)

# 결과 출력
print(result)
result
```
